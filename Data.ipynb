```python
import json
import pandas as pd
```

```python
#Load dataset to a data frame
data = pd.read_csv(r'Leetcode.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
#Fetch topics of all questions
topics_all_questions = [data.loc[i]['questionTags'].split(', ') for i in data.index]
```

```python
topics_all_questions
```

```python
#Find the list of topics of all questions in alphabetical order
topics = sorted(list(set([item for sublist in topics_all_questions for item in sublist])))
```

```python
topics
```

```python
percentage = [data.loc[i]['acceptancePercentage'].translate({ord('%'): None}) for i in data.index]
```

```python
percentage
```

```python
# Initialize lists for questions data as well as titles
full_data = list()
question_titles = list()
```

```python
for i in data.index:
    # Append question title and the index of the question
    question_titles.append((data.loc[i]['questionName'].strip(), i, data.loc[i]['questionURL'].strip()))
    # Add list of topics of the movies (1/0) to question data
    question_data = [1 if topic in data.loc[i]['questionTags'].split(', ') else 0 for topic in topics]
    # Add acceptancePercentage of the question to the question data
    question_data.append(int(float(percentage[i])))
    # Add record of question to main data list
    full_data.append(question_data)
```

```python
full_data
```

```python
question_titles
```

```python
# Create JSON files for data and questions titles for faster load to the Recommmender
data_dump = r'data.json'
titles_dump = r'titles.json'
with open(data_dump, 'w+', encoding='utf-8') as f:
    json.dump(full_data, f)
with open(titles_dump, 'w+', encoding='utf-8') as f:
    json.dump(question_titles, f)
```
